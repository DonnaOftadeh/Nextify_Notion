## Quick project info form

Run this cell to render the form, enter your details, and click **Submit**.
Your responses will be stored in a `project_info` dictionary for reuse later in the notebook.


```python
import ipywidgets as widgets
from IPython.display import display, clear_output

name_input = widgets.Text(description='Name', placeholder='Your name', style={'description_width': 'initial'})
email_input = widgets.Text(description='Email', placeholder='you@example.com', style={'description_width': 'initial'})
goal_input = widgets.Textarea(
    description='Project goal',
    placeholder='What do you want the agent to help with?',
    layout=widgets.Layout(width='500px', height='120px'),
    style={'description_width': 'initial'},
)
submit_btn = widgets.Button(description='Submit', button_style='success', tooltip='Submit your project info')
output = widgets.Output()

project_info = {}

def handle_submit(_):
    project_info['name'] = name_input.value.strip()
    project_info['email'] = email_input.value.strip()
    project_info['goal'] = goal_input.value.strip()

    with output:
        clear_output()
        print('Thanks for the details! Here is what you submitted:')
        print(f"  Name: {project_info['name'] or '(not provided)'}")
        print(f"  Email: {project_info['email'] or '(not provided)'}")
        print('  Project goal:')
        if project_info['goal']:
            print(f"    {project_info['goal']}")
        else:
            print('    (not provided)')
        print('
You can reuse the `project_info` dictionary in later cells.')

submit_btn.on_click(handle_submit)
form = widgets.VBox([name_input, email_input, goal_input, submit_btn], layout=widgets.Layout(gap='10px'))
display(form, output)
```
